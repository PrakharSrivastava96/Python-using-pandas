### Step 1. Import the necessary libraries

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data=pd.read_csv("wind.data")
data['Date_index']=pd.to_datetime(data.Yr*10000+data.Mo*100+data.Dy,format="%y%m%d")
del data['Yr']
del data['Mo']
del data['Dy']
data.head()
```

### Step 4. Year 2061 is seemingly imporoper. Convert every year which are < 70 to 19XX instead of 20XX.

```python
data['Date_index']=np.where(data.Date_index.dt.year>2000,data.Date_index-pd.offsets.DateOffset(years=100),data.Date_index)
data.head()
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data1=data.set_index('Date_index')
data1.head()
```
