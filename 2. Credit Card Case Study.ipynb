```python
import pandas as pd
import numpy as np
```

```python
customer_Acq=pd.read_csv('Customer Acqusition.csv')
customer_Acq.head(2)
```

```python
repayment=pd.read_csv('Repayment.csv')
repayment.head(2)
```

```python
spend=pd.read_csv('spend.csv')
spend.head(2)
```

```python
credit_card=pd.merge(customer_Acq,spend,on='Customer',how='inner')
credit_card.head(2)
```

```python
credit_card=pd.merge(credit_card,Repayment,on='Customer',how='inner')
credit_card.head(2)
```

```python
credit_cardf=credit_card.drop(['Sl No:','SL No:','Unnamed: 4','No'],axis=1)
credit_cardf.rename(columns={'Month_x':'spent_month','Month_y':'repay_month','Amount_x':'spent_amt','Amount_y':'repay_amt'},inplace=True)
```

```python
credit_cardf.head(2)
```

```python
credit_cardf.dtypes
```

```python
credit_cardf['spent_month']=pd.to_datetime(credit_cardf.spent_month)
```

```python
credit_cardf['repay_month']=pd.to_datetime(credit_cardf.repay_month)
```

```python
credit_cardf.head(2)
```

##  a. In case age is less than 18, replace it with mean of age values

```python
credit_cardf.Age.loc[credit_cardf.Age<18]=credit_cardf.Age.mean()
```

## b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit

```python
credit_cardf.spent_amt.loc[credit_cardf.spent_amt>credit_cardf.Limit]=0
```

```python
credit_cardf.spent_amt.replace(0,credit_cardf.Limit/2,inplace=True)
```

```python
credit_cardf.head(2)
```

##  c. Incase the repayment amount is more than the limit, replace the repayment with the limit. 

```python
credit_cardf.repay_amt.loc[credit_cardf.repay_amt>credit_cardf.Limit]=0
```

```python
credit_cardf.repay_amt.replace(0,credit_cardf.Limit,inplace=True)
```

##  c. & d. What is the average monthly spend by customers and the average monthly repayment by customers?

```python
credit_cardf['month'] = credit_cardf.spent_month.dt.month
monthly_data=credit_cardf.groupby(by=['month'])[['spent_amt','repay_amt']].mean()
```

```python
credit_cardf.head(2)
```

```python
monthly_data
```

##   e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? (Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment – Monthly spend. Interest is earned only on positive profits and not on negative amounts

```python
monthly_data['monthly_profit']=monthly_data.repay_amt-monthly_data.spent_amt
```

```python
bank_profit=monthly_data[monthly_data.monthly_profit>1]
```

```python
bank_profit['monthly_bank_profit']=bank_profit.monthly_profit*(2.9/100)
```

```python
bank_profit
```

##   f. What are the top 5 product types?

```python
top_5_product_types=credit_cardf.groupby(['Type'])[['spent_amt']].count().add_prefix("Total_of ")
```

```python
top_5_product_types.head(5)
```

```python
top_5_product_types.sort_values(by='Total_of spent_amt',ascending=False).head(5)
```

##   g. Which city is having maximum spend?

```python
max_spend=credit_cardf.groupby(['City'])[['spent_amt']].sum()
```

```python
max_spend.sort_values(by='spent_amt',ascending=False).head(1)
```

## h. Which age group is spending more money?

```python
bins=[19,35,50,65,80]
group_names=['young','Mid_age','old','Very_old']
credit_cardf['Age_cat']=pd.cut(credit_cardf['Age'],bins,labels=group_names)
```

```python
credit_cardf.head(2)
```

```python
max_spent_agegroup=credit_cardf.groupby(['Age_cat'])[['spent_amt']].sum()
```

```python
max_spent_agegroup.sort_values(by='spent_amt',ascending=False).head(2)    ## age 35-50 is spending more
```
